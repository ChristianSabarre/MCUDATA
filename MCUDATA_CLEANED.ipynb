```python
import pandas as pd

summary_list = []

for phase in sorted(df_clean['mcu_phase'].unique()):
    phase_movies = df_clean[df_clean['mcu_phase'] == phase]
    
    summary_list.append({
        'mcu_phase': phase,
        'avg_budget': phase_movies['production_budget'].mean(),
        'med_budget': phase_movies['production_budget'].median(),
        'avg_opening_weekend': phase_movies['opening_weekend'].mean(),
        'med_opening_weekend': phase_movies['opening_weekend'].median(),
        'avg_domestic_box_office': phase_movies['domestic_box_office'].mean(),
        'med_domestic_box_office': phase_movies['domestic_box_office'].median(),
        'avg_international_box_office': phase_movies['international_box_office'].mean(),
        'med_international_box_office': phase_movies['international_box_office'].median(),
        'avg_worldwide_box_office': phase_movies['worldwide_box_office'].mean(),
        'med_worldwide_box_office': phase_movies['worldwide_box_office'].median(),
        'med_imdb_rating': phase_movies['imdb_rating'].median(),
        'avg_runtime': phase_movies['movie_duration'].mean(),
        'med_runtime': phase_movies['movie_duration'].median(),
        'med_tomato_meter': phase_movies['tomato_meter'].median(),
        'med_audience_score': phase_movies['audience_score'].median(),
        'movie_count': len(phase_movies)
    })

# Convert to DataFrame
phase_summary_df = pd.DataFrame(summary_list)

# Save to CSV
phase_summary_df.to_csv('mcu_phase_summary.csv', index=False)

print("CSV saved as 'mcu_phase_summary.csv'")
```

```python
summary = []

for phase in sorted(df_clean['mcu_phase'].unique()):
    phase_movies = df_clean[df_clean['mcu_phase'] == phase]

    mean_revenue = phase_movies['worldwide_box_office'].mean()
    median_revenue = phase_movies['worldwide_box_office'].median()
    total_revenue = phase_movies['worldwide_box_office'].sum()

    mean_budget = phase_movies['production_budget'].mean()
    
    # Budget-to-revenue ratio per movie
    phase_movies = phase_movies[phase_movies['production_budget'] > 0]  # avoid divide by 0
    ratios = phase_movies['worldwide_box_office'] / phase_movies['production_budget']
    
    mean_ratio = ratios.mean()
    median_ratio = ratios.median()

    summary.append({
        'mcu_phase': phase,
        'mean_revenue': mean_revenue,
        'median_revenue': median_revenue,
        'total_revenue': total_revenue,
        'mean_budget': mean_budget,
        'mean_return_ratio': mean_ratio,
        'median_return_ratio': median_ratio
    })

summary_df = pd.DataFrame(summary)
summary_df.to_csv(r'C:\Users\ojt_opsdev_christian\Desktop\mcu\mcu_phase_revenue_analysis.csv', index=False)
print("CSV saved as 'mcu_phase_revenue_analysis.csv'")
```
